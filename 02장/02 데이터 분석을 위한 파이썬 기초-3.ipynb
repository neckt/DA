### **1. 조건문**

```python
#코드 2-9
a = 5

if a == 5:
  print('Right!')
  print('a is 5')

if a == 3:
  print('Right!')
  print('a is 3')

if a != 3:
  print('Right!')
  print('a is not 3')
```

### **2. 반복문**

```python
#코드 2-12
a = 5
i = 1

#9번 반복하기
while i <= 9:
  print(str(a) + ' X ' + str(i) + ' = ' + str(i*a))
  i += 1
print('파이썬으로 구구단 5단을 계산할 수 있다!')
```

```python
#코드 2-13
a = 5

for i in range(1,10):
  print(str(a) + ' X ' + str(i) + ' = ' + str(i*a))
print('while 조건문을 for 조건문으로 바꾸어 사용할 수 있다!')
```
