### **3. 파이썬으로 엑셀 파일 작성하기**

```python
#코드 3-12
wb = openpyxl.Workbook()
```

```python
#코드 3-13
#새 워크시트 생성하기
wb.create_sheet('Sheet2')
#시트 이름 출력하기
print(wb.sheetnames)
```

```python
#코드 3-14
sheet1 = wb['Sheet']
sheet2 = wb['Sheet2']

sheet1.title = '캐릭터 명단'
sheet2.title = '인기도 조사'

print(wb.sheetnames)
```

```python
#코드 3-15
sheet2['B1'] = '인기도 조사결과'
print(sheet2['B1'].value)
```

```python
#코드 3-16
copysheet = wb.copy_worksheet(sheet2)
print(wb.sheetnames)

#복제 시트의 이름 변경하기
copysheet.title = 'copy'

print(wb.sheetnames)
print(copysheet['B1'].value)
```

```python
#코드 3-17
del wb['인기도 조사']
print(wb.sheetnames)
```

```python
#코드 3-18
wb.save('complete.xlsx')
```
